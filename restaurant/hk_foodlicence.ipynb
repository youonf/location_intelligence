```python
import pandas as pd
from tqdm import tqdm
```

```python
data = json.loads(text)
df = pd.DataFrame(columns=
['company_name',
 'district',
 'address',
 'license_no',
 'record_count',
'1',
'2',
'3',
'4',
'5',
'6',
'7',
])

license_type = {'Note 1': 1,
                'Note 2': 2,
                'Note 3': 3,
                'Note 4': 4,
                'Note 5': 5,
                'Note 6': 6,
                'Note 7': 7}

for i in tqdm(range(232)):
    record = 50*i
    text = requests.get('https://www.fehd.gov.hk/english/licensing/getData.php?lang=en&from='+str(record)+'&licenseType=General%20Restaurant%20Licence').text
    data = json.loads(text)
    for j in range(len(data)):
        license_dict = {}
        if '<' in data[j]['companyName']:
            for l in license_type.keys():
                if l in data[j]['companyName']:
                    license_dict[str(license_type[l])] = 1
            license_dict['company_name'] = data[j]['companyName'].split('<')[0]
        else:     
            license_dict['company_name'] = data[j]['companyName']
        license_dict['district'] = data[j]['district']
        license_dict['address'] = data[j]['address']
        license_dict['license_no'] = data[j]['licenseNo']
        license_dict['record_count'] = data[j]['recordCount']

        df = df.append(license_dict, ignore_index=True)
```
